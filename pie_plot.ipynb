```python
"""#Pie plot :
It's univariante plot
gives %wise distribution
circle divided into a series of segments.
Each segment represents a particular category.
"""
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t = sns.load_dataset('titanic')
t.head(3)
```

```python
t['class']
```

```python
# to get categorywise count use value_counts()
cp = t['class'].value_counts()
cp
```

```python
plt.pie(cp.values)
```

```python
# to show %
plt.pie(cp.values,autopct='%.2f%%')
```

```python
#to show labels
plt.pie(cp.values,labels=cp.index,autopct='%.2f%%')
```

```python
plt.pie(cp.values,labels=cp.index,autopct='%.2f%%',colors=['olivedrab', 'rosybrown', 'gray'])
```

```python
#Hatch slices
#Pass a list of hatch patterns to hatch to set the pattern of each slice.
plt.pie(cp.values,labels=cp.index,colors=['olivedrab', 'rosybrown', 'gray'],hatch=['**O', 'oO', 'O.O']);
```

```python
#Swap label and autopct text positions
#Use the labeldistance and pctdistance parameters to position the labels and autopct text respectively.

plt.pie(cp.values,labels=cp.index, autopct='%1.1f%%',
       pctdistance=1.25, labeldistance=.6);
```

```python
#To explode
plt.pie(cp.values,labels=cp.index, autopct='%1.1f%%',explode=[0,0.2,0]);
```

```python
#add a drop-shadow using shadow
plt.pie(cp.values,labels=cp.index, autopct='%1.1f%%',explode=[0,0.2,0],shadow=True);
```
