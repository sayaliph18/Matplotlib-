```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = sns.load_dataset('diamonds')
d.head()
```

```python
#Line plot: Plot y versus x as lines and/or markers.
#It is univariate,bivariate and multivariate as well
#used to express a relationship between two variables.
plt.plot?
```

```python
plt.plot(d.carat) # index on x-axis and values on y-axis
```

```python
#x vs y plot
plt.plot(d.x,d.y)
```

```python
# to reduce size of plot
plt.figure(figsize=(6,4))
plt.plot(d.x,label='x column')
plt.plot(d.z,label='z column')
plt.legend() #to display label in the box
plt.title('Sample line plot')
plt.xlabel('Index')
plt.ylabel('Input Data')
```

```python
# to separate both graphs
plt.figure(figsize=(6,4))
plt.plot(d.x,label='x column')
plt.legend() #to display label in the box
plt.title('Sample line plot_1')
plt.xlabel('Index_1')
plt.ylabel('Input Data_1')



plt.figure(figsize=(6,4))
plt.plot(d.z,label='z column')
plt.legend() #to display label in the box
plt.title('Sample line plot_2')
plt.xlabel('Index_2')
plt.ylabel('Input Data_2')
```

```python
# Sample data (time periods and corresponding values)
time_periods = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
values = [100, 120, 110, 130, 140, 150]
```

```python
#creating line plot
plt.plot(time_periods,values,marker='o',linestyle='-')
#adding labels and title
plt.xlabel('Time Period')
plt.ylabel('Values')
plt.title('Trend analysis over time')

#Display the plot
plt.grid(True) #Add a grid for better visualization
plt.show()
```

```python
# Sample data for 2022 and 2023
years = ['2022', '2023']
values_2022 = [100, 120, 110, 130, 140, 150]
values_2023 = [110, 130, 125, 145, 155, 165]

#create a line plot for 2022
plt.plot(values_2022,label=2022,marker='o',linestyle='-')

#create a line plot for 2023
plt.plot(values_2023,label=2023,marker='o',linestyle='-')

#Adding labels and a title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trand analysis for 2022 and 2023')

#Display a legend to differentiate the years
plt.legend()

#Display the plot
plt.grid(True)
plt.show()
```
