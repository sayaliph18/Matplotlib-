```python
"""
Scatterplot:
-A scatter plot of y vs. x with varying marker size and/or color
-graphs pairs of numerical data, with one variable on each axis, to look for a relationship between them
-to explore the relationship between two variables
-to display the relationship between two variables and observe the nature of the relationship.

"""
```

```python
import numpy as np
age = np.random.randint(2,80,50)
age
```

```python
import matplotlib.pyplot as plt
```

```python
new_age = np.append(age,[-10,-20,150,175,210])
new_age
```

```python
## use scatter plot to detect outlier

plt.scatter(range(55),new_age);
```
