```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = sns.load_dataset('diamonds')
d.head()
```

```python
days = ['monday','tuesday','wednesday','thursday']
count = [24,45,13,22]
```

```python
#Barplot: Category (on x-axis) vs count(on y-axis) plot
"""
- It is bivariant plot
shows the relationship between a numeric and a categoric variable.
type of data visualization used to represent data in the form of rectangular bars. 
The height of each bar represents the value of a data point, and the width of each bar represents the category of the data.
"""
plt.bar?
```

```python
plt.bar(days,count)
plt.savefig('bar.jpeg')  #to save the plot
```

```python
plt.bar(days,count,width=.4,alpha=.3,color=['r','y','m','c'])
```

```python
plt.bar(days,count,width=.4,alpha=.3,color=['r','y','m','c'])
#To chenage x label
plt.xticks(range(4),['Mon','Tues','Wed','Thurs'])
plt.show() #to remove unneccessary info
```

```python
plt.bar(days,count,width=.4,alpha=.3,color=['r','y','m','c'])
#To chenage x label
plt.xticks(range(4),['Mon','Tues','Wed','Thurs'],rotation=45)
plt.show() #to remove unneccessary info
```
