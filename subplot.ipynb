```python
"""
Subplot: 
-The two integer arguments to this function specify the number of rows and columns of the subplot grid
- grid of smaller axes, with each axis being a plot that is plotted together within a single figure

"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sale = [240,120,450,300,100]
state = ['Maharashtra','Haryana','Punjab','Karnataka','Uttar Pradesh']
```

```python
age = np.random.randint(2,80,50)
age
```

```python
new_age = np.append(age,[-10,-20,150,175,210])
new_age
```

```python
plt.subplot(121)
plt.subplot(122)
```

```python
plt.subplot(221)
plt.subplot(222)
plt.subplot(223)
plt.subplot(224)
```

```python
plt.subplot(321)
plt.subplot(322)
plt.subplot(323)
plt.subplot(324)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.barh(state,sale,color='red')
plt.title('State vs Sale plot 1')

plt.subplot(222)
plt.boxplot(age)
plt.title('Boxplot Age')

```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.barh(state,sale,color='red')
plt.title('State vs Sale plot 1')

plt.subplot(222)
plt.boxplot(age)
plt.title('Boxplot Age')

plt.subplot(223)
plt.barh(state,sale,color='g')
plt.title('State vs Sale plot 2')

```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.barh(state,sale,color='red')
plt.title('State vs Sale plot 1')

plt.subplot(222)
plt.boxplot(age)
plt.title('Boxplot Age')

plt.subplot(223)
plt.barh(state,sale,color='g')
plt.title('State vs Sale plot 2')

plt.subplot(224)
plt.hist(new_age,color='y')
plt.title('Histogram newAge')
```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.barh(state,sale,color='red')
plt.title('State vs Sale plot 1')

plt.subplot(222)
plt.boxplot(age)
plt.title('Boxplot Age')

plt.subplot(223)
plt.barh(state,sale,color='g')
plt.title('State vs Sale plot 2')

plt.subplot(224)
plt.hist(new_age,color='y')
plt.title('Histogram newAge')

plt.suptitle('Combining 4 plots')
plt.tight_layout()
#save fig
plt.savefig('sample.jpg',dpi=300)
plt.show()
```
