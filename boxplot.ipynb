```python
import numpy as np
age = np.random.randint(2,80,50)
age
```

```python
plt.boxplot(age);
```

```python
#Adding Elements to the Box Plot 
plt.boxplot(age)
plt.ylabel('age')
plt.title('Boxplot of Age')
plt.show()
```

```python
#lets  add some outliers
new_age = np.append(age,[-10,-20,150,175,210])
new_age
```

```python
#Notch : The notches represent the confidence interval (CI) around the median 
plt.boxplot(new_age,notch=True)
plt.ylabel('age')
plt.title('Boxplot of Age')
plt.show()
```

```python
#vert: This parameter is set to boolean values False or True for horizontal and vertical plot respectively:
plt.boxplot(new_age,vert=False)
plt.ylabel('age')
plt.title('Boxplot of Age')
plt.show()
```

```python
#patch_artist: This parameter is set to boolean False by default and produces boxes with Line2D artist. 
#If set to True produces a plot with Patch artists
plt.boxplot(new_age,patch_artist=True)
plt.ylabel('age')
plt.title('Boxplot of Age')
plt.show()
```

```python
#manage: This parameter if set to boolean True by default and the tick locations and labels are adjusted to match the boxplot positions.
#If set to False, this happens
plt.boxplot(new_age,manage_ticks=False)
plt.ylabel('age')
plt.title('Boxplot of Age')
plt.show()
```

```python
#showmeans: This parameter if set to boolean value True, displays plot mean:
plt.boxplot(new_age,showmeans=True)
plt.ylabel('age')
plt.title('Boxplot of Age')
plt.show()
```
